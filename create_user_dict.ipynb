```python
from surprise import Dataset
import numpy as np
import pandas as pd
```

```python
data = Dataset.load_builtin('ml-1m')
df = pd.DataFrame(data.raw_ratings, columns = ['UserId', 'MovieId', 'Rating',  'Timestamp'], dtype='int32')
```

```python
users_dict = {int(user) : [] for user in set(df["UserId"])}
```

```python
df_gen = df.iterrows()
users_dict_for_history_len = {user : [] for user in set(df["UserId"])}
for data in df_gen:
    users_dict[int(data[1]['UserId'])].append((int(data[1]['MovieId']), int(data[1]['Rating'])))
    if data[1]['Rating'] >= 4:
        users_dict_for_history_len[data[1]['UserId']].append((int(data[1]['MovieId']), int(data[1]['Rating'])))

users_history_lens = [len(users_dict_for_history_len[u]) for u in set(df["UserId"])]
```
